```python
import cv2
from PIL import Image
import numpy as np
from utils import get_limits
```

```python
yellow = [0, 255, 255]  # yellow in BGR colorspace 


cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()

    

    hvsImage = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    lower_yellow, upper_yellow= get_limits(color=yellow)

    lower_blue = np.array([110,50,50])
    upper_blue = np.array([130,255,255])
    mask_yellow = cv2.inRange(hvsImage, lower_yellow, upper_yellow)
    mask_blue = cv2.inRange(hvsImage, lower_blue, upper_blue)


    mask_yellow = Image.fromarray(mask_yellow)
    bbox_yellow = mask_yellow.getbbox()

    mask_blue = Image.fromarray(mask_blue)
    bbox_blue = mask_blue.getbbox()

    print(bbox_yellow)
    if bbox_yellow is not None:
        x1, y1, x2, y2 = bbox_yellow
        
        cv2.rectangle(frame, (x1,y1), (x2,y2), (0,255,0), 5)
        

    if  bbox_blue is not None:
        a1, s1, a2, s2 = bbox_blue
        cv2.rectangle(frame, (a1,s1), (a2,s2), (0,0,255), 5)
    
    
    cv2.imshow('window', frame)
    # cv2.imshow('frame', mask)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()


```
